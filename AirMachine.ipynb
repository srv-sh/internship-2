```python
# drop null value from column
for i in data.columns:
    if data[i].isna().sum() == len(data):
        data.drop(i,axis=1,inplace=True)
```

```python
# drop null value from row
ar = []
for i in range(data.shape[0]):
    if data.iloc[i].isna().sum() == data.shape[1]:
        ar.append(i)
data = data.drop(ar)
```

```python
nw = data.copy()
```

```python
def month_wise(column):
    month = nw[column]
    new_header = month.iloc[0] #grab the first row for the header
    month = month[1:] #take the data lessheader row
    month.columns = new_header #set the header row as the df header
    return month
```

```python
df = pd.DataFrame()
col= nw.columns
for i in range(0,len(col),5):
    month = month_wise(col[i:i+5])
    df = pd.concat([df,month])
    
```

```python
df.shape
```

```python
#filtering negative rows
for cols in df.columns.tolist()[1:4]:
    df= df.loc[df[cols] > 0]
```

```python
df.shape
```

### Arithmetic operation 

```python
df['Total production (ton)'] = df['production_1'] + df['production_2'] + df['production_3']
df.head(2)
```

```python
# create 3 columns to store percentage value
df['production_1_percentage'] = 0
df['production_2_percentage'] = 0
df['production_3_percentage'] = 0
```

```python
df
```

```python
# store appropiate percentage value
for idx in range(df.shape[0]):
    if df['Total production (ton)'].iloc[idx] == 0 or df['production_1'].iloc[idx] == 0:
        df['production_1_percentage'].iloc[idx] = 0
    else:
        df['production_1_percentage'].iloc[idx] = df['production_1'].iloc[idx] / df['Total production (ton)'].iloc[idx]
        
    if df['Total production (ton)'].iloc[idx] == 0 or df['production_2'].iloc[idx] == 0:
        df['production_2_percentage'].iloc[idx] = 0
    else:
        df['production_2_percentage'].iloc[idx] = df['production_2'].iloc[idx] / df['Total production (ton)'].iloc[idx]
        
    if df['Total production (ton)'].iloc[idx] == 0 or df['production_3'].iloc[idx] == 0:
        df['production_3_percentage'].iloc[idx] = 0
    else:
        df['production_3_percentage'].iloc[idx] = df['production_3'].iloc[idx] / df['Total production (ton)'].iloc[idx]
        
        
        
        
```

```python
# reset the index values
df =df.reset_index(drop=True)
```

```python
def distance(point1 , point2):
    dist = np.linalg.norm(point1 - point2)
    return dist
```

```python
data_new['outliner'] =0
out_df = pd.DataFrame()
```

```python
for i in range(len(centroids)):
    dist_list = []
    data= data_new[data_new['cluster']==i]
    for j in range(len(data)):
        point2= centroids[i]
        point1 = np.array(data.iloc[j].to_list()[5:-2])
        dist =  distance(point1 , point2)
        dist_list.append(dist)
    mean = np.array(dist_list).mean()
    std = np.array(dist_list).std()
    maxi =np.array(dist_list).max()
    mini = np.array(dist_list).min()
    print('for cluster {}'.format(i))
    print('mean :{}|std :{}|maxi:{}|mini:{}'.format(mean,std,maxi,mini))
    for k in range(len(data)):
        point2= centroids[i]
        point1 = np.array(data.iloc[k].to_list()[5:-2])
        dist =  distance(point1 , point2)
        if dist <= mean+(2*std):
            pass
        else:
            print("{} > {}".format(dist,mean+(2*std)))
            data['outliner'].iloc[k] = 1
            #test['cluster'].iloc[i] =result[0]
    out_df = pd.concat([out_df,data])
```
